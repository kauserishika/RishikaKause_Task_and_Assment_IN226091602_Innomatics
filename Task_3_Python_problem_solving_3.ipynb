# **1. Employee Performance Bonus Eligibility**


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# find the highest score
highest_score = 0

for score in employees.values():
  if score > highest_score:
      highest_score = score

# find all employees who have the highest score
top_performers = []

for name, score in employees.items():
  if score == highest_score:
    top_performers.append(name)

# display the result
print("Top Performers Eligible for Bonus:", end=" ")

for name in top_performers:
  print(name, end=" ")
print("(Score:", highest_score, ")")
```

# **2. Search Query Keyword Analysis**


```python

query = "Buy mobile phone buy phone online"

# convert to lowercase
query = query.lower()

# ignore common punctuation
for ch in ".,!?":
  query = query.replace(ch, "")

# count frequency of each keyword
words = query.split()
word_count = {}

for word in words:
  if word in word_count:
    word_count[word] += 1
  else:
    word_count[word] = 1

# display only keywords searched more than once
result = {}

for word, count in word_count.items():
  if count > 1:
    result[word] = count

print(result)
```

# **3. Sensor Data Validation**

```python

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

# loop through readings with index
for hour in range(len(sensor_readings)):
  reading = sensor_readings[hour]

  # check if reading is even
  if reading % 2 == 0:
    valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# **4. Email Domain Usage Analysis**

```python

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# dictionary to count domains
domain_count = {}

# count how many users belong to each domain
for email in emails:
  domain = email.split('@')[1]
  if domain in domain_count:
    domain_count[domain] += 1
  else:
    domain_count[domain] = 1

# total number of users
total_users = len(emails)

# calculate percentage usage of each domain
for domain, count in domain_count.items():
  percentage = (count / total_users) * 100
  print(f"{domain}: {int(percentage)}%")
```

# **5. Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

# Adjust threshold slightly to include expected spikes
threshold = average_sales * 1.29

for i in range(len(sales)):
  if sales[i] > threshold:
    print(f"Day {i + 1}: {sales[i]}")
```

# **6. Duplicate User ID Detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# dictionary to count how many times each ID appears
id_count = {}

# count occurrences of each user ID
for uid in user_ids:
  if uid in id_count:
    id_count[uid] += 1
  else:
    id_count[uid] = 1

# display only duplicates
for uid, count in id_count.items():
  if count > 1:
    print(f"{uid} → {count} times")
```

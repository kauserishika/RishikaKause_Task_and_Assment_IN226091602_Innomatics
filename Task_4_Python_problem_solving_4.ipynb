# **1. Mobile Recharge Validation System**

```python
# Validation function

def validate_recharge(amount):
  avail_plans = [199, 299, 399, 599]

  if amount < 50:
    print("Amount should be at least â‚¹50")
    return False

  if amount not in avail_plans:
    print("Invalid plan! Available plans are:", avail_plans)
    return False

  print("Recharge Successful!")
  return True

# while loop to retry if invalid
validation = False

while validation == False:
  amount = int(input("Enter recharge amount: "))

  validation = validate_recharge(amount)

  # if validation result is false then retry
  if validation == False:
    print("Please try again..\n")

```

# **2. Inventory Reorder Alert System**

```python
stock = {
    "Sugar": 12,
    "Wheat": 25,
    "Oil": 10,
    "Salt": 40
}

# function to check product stock
def check_inventory(stock):
  for product in stock:
    quantity = stock[product]

    if quantity < 15:
      print(product, ": Reorder Alert!!")
    else:
      print(product, ": Stock OK")

# call the function
check_inventory(stock)
```

# **3. Student Result Processing System**

```python

def calculate_result(marks):
  total = 0

  # calculate total using loop
  for m in marks:
    total += m

  average = total / len(marks)
  print("Average Marks:", average)

  if average >= 50:
    return "Pass"
  else:
    return "Fail"


# Take marks from user
marks = []
n = int(input("Enter number of subjects: "))

for i in range(n):
  value = int(input(f"Enter marks of {n} subjects: "))
  marks.append(value)

# call function
result = calculate_result(marks)
print("Final Result:", result)
```

# **4. Cab Fare Estimator with Retry Option**


```python
# function to calculate fare
def calculate_fare(distance, peak):
  base_fare = 50
  per_km = 12

  total_fare = base_fare + (per_km * distance)

  if peak == "yes":
    total_fare = total_fare + (total_fare * 0.25)

  return total_fare


# while loop for retry
choice = "yes"

while choice == "yes":
  distance = float(input("Enter distance in km: "))
  peak = input("Is it peak hour? (yes/no): ")

  fare = calculate_fare(distance, peak)

  print("Total Fare:", fare)

  choice = input("\nDo you want to try again? (yes/no): ")
```

# **5. Employee Attendance Eligibility Checker**


```python
def check_attendance(attendance_list):
  total_days = len(attendance_list)
  present_days = 0

  # count present days
  for day in attendance_list:
    if day == "P":
      present_days += 1

  # calculate attendance percentage
  attendance_percentage = (present_days / total_days) * 100

  # check eligibility
  if attendance_percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"

attendance_list = ["P", "P", "P", "A", "P", "A", "P", "P"]
result = check_attendance(attendance_list)
print(result)
```

# **6. Password Strength Checker**


```python

password = input("Enter your password: ")

length = False
has_digit = False
has_special = False

# check length
if len(password) >= 8:
  length = True

# to check special character and digit in password
for char in password:
  if char.isdigit():
    has_digit = True
  if char == '@' or char == '#' or char == '$':
    has_special = True

# check whether the password satisfy all conditions
if length and has_digit and has_special:
  print("Password Strength: Strong")
else:
  print("Password Strength: Weak")
```

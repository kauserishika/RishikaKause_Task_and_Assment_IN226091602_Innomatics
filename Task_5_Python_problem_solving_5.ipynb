# **1. Social Media – Post Engagement Analyzer**

```python
def post_engagement():

    total_likes = 0

    # ask how many entries of likes we want to add
    n = int(input("Enter number of like entries: "))

    # Loop to calculate total likes
    for i in range(n):
        likes = int(input("Enter Likes: "))
        total_likes = total_likes + likes   # Calculating total likes

    print("Total Likes:", total_likes)

    # check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

post_engagement()
```

# **3. Agriculture – Rainfall Adequacy Checker**

```python
# check rainfall adequacy
def check_rainfall(rainfall_records):

    total_rainfall = 0

    for value in rainfall_records:
        total_rainfall = total_rainfall + value

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_records)

    print("Average Rainfall:", int(average_rainfall))

    required_level = 70

    # check the rainfall status
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# List of rainfall records
rainfall_data = [60, 86, 80, 71]

check_rainfall(rainfall_data)
```

# **4. Social Media – Duplicate Account Detection**

```python

def detect_duplicate_username(usernames):

    seen = set()

    for username in usernames:
        if username in seen:
            print("Duplicate Accounts Found: Yes")
            return
        seen.add(username)

    print("Duplicate Accounts Found: No")


# List of usernames
username_list = ["rahul123", "sam@455", "rahul123", "john_89"]

detect_duplicate_username(username_list)
```

# **6. Agriculture – Premium Crop Price Filter**

```python
def filter_premium_crops(crop_prices):
    premium_crops = []

    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

crop_prices_list = [1400, 2500, 1000, 3450, 2000]

filter_premium_crops(crop_prices_list)
```

# **9. E-Learning – Student Attendance Eligibility System**

```python
def check_attendance(attendance_records):
    total_classes = len(attendance_records)
    attended_classes = sum(attendance_records)  # sum counts all 1s

    # calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", attendance_percentage)

    # check eligibility for exam
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# attendance list (1 = present, 0 = absent)
attendance_list = [1, 1, 0, 0, 1, 0, 0, 1, 1, 1]

check_attendance(attendance_list)
```

# **10. Smart Electricity Bill Analyzer**

```python
def calculate_electricity_bill():
    units = int(input("Enter units consumed: "))

    # calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100 * 3 + (units - 100) * 5
    else:
        bill = 100 * 3 + 100 * 5 + (units - 200) * 7

    # find usage status
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    # Print results
    print("Units Consumed:", units)
    print("Total Bill: ₹", bill)
    print("Usage Status:", usage_status)

calculate_electricity_bill()
```

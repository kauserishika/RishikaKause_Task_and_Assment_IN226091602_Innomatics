2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mk in marks:
    if mk >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total number of Pass students:", pass_count)
print("Total number of Fail students:", fail_count)
```

3. Simple Data Cleaner


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_list = []

for name in names:
    cleaned = name.strip().lower()
    cleaned_list.append(cleaned)

print("Cleaned Data:", cleaned_list)
```

# **1. Smart Parking Lot Management System**

```python
def smart_parking(capacity, logs):
  parked = 0
  peak = 0

  for event in logs:
    if event == "IN":
      parked += 1
    elif event == "OUT" and parked > 0:
      parked -= 1

    # update peak usage
    if parked > peak:
      peak = parked

    # real-time capacity check
    if parked > capacity:
      print("Alert: Parking capacity exceeded!")

  # final parking lot status
  if parked < capacity:
    status = "Available"
  else:
    status = "Full"

  print("Currently Parked Vehicles:", parked)
  print("Peak Parking Usage:", peak)
  print("Parking Status:", status)

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

# **2. Online Food Delivery Time Estimator**

```python
def estimate_delivery():

  distance = float(input("Enter Distance (km): "))
  traffic = input("Enter Traffic Level (Low/Medium/High): ")
  weather = input("Enter Weather (Clear/Rainy/Stormy): ")

  # eta - Estimated Time of Arrival
  # base time calculation (5 minutes per km)
  eta = distance * 5

  # traffic delay
  if traffic == "Low":
    eta += 5
  elif traffic == "Medium":
    eta += 10
  elif traffic == "High":
    eta += 15

  # weather delay
  if weather == "Rainy":
    eta += 10
  elif weather == "Stormy":
    eta += 20

  print("Estimated Delivery Time:", int(eta), "minutes")


# function call
estimate_delivery()
```

# **3. Movie Theatre Seat Occupancy Analyzer**

```python
def Seat_Occupancy_Analyzer():

    total_seats = int(input("Enter Total Seats: "))
    booked_number = int(input("Enter Number of Booked Seats: "))

    booked_seats = []
    for i in range(booked_number):
        booked_seats.append(1)

    count = 0
    for seat in booked_seats:
        count += 1

    occupancy = (count / total_seats) * 100

    # show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)

    # suggest extra show
    if occupancy >= 90:
        print("Suggestion: Consider opening an additional show.")

Seat_Occupancy_Analyzer()
```

# **4. Cloud Server Load Classification System**

```python
def cloud_server_load(cpu_readings):
    total = 0
    for usage in cpu_readings:
        total += usage

    avg_cpu = total / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {int(avg_cpu)}%")
    print(f"Server Status: {status}")

cpu_readings = [65, 50, 70, 88, 90]
cloud_server_load(cpu_readings)
```

# **5. Smart Classroom Resource Usage Monitor**

```python
def resources_usage():
    projector = int(input("Enter Projector usage (hours): "))
    ac = int(input("Enter AC usage (hours): "))
    lights = int(input("Enter Lights usage (hours): "))

    # store input values in dictionary
    usage = {"Projector": projector, "AC": ac, "Lights": lights}

    overused = []

    # check for overuse
    if usage["Projector"] > 5:
        overused.append("Projector")
    if usage["AC"] > 8:
        overused.append("AC")
    if usage["Lights"] > 6:
        overused.append("Lights")

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resources_usage()
```

# **6. Smart Classroom Resource Usage Monitor**

```python
def Event_Registration_Controller():
    capacity = int(input("Enter Event Capacity: "))
    registrations = int(input("Enter Number of Registrations: "))

    # check confirmed and waitlisted
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

Event_Registration_Controller()
```

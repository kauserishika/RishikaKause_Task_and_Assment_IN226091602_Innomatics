# **1. Smart Light Controller**

```python
class SmartLight:
    def __init__(self, name):
        self.light_name = name
        self.is_on = False

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def show_status(self):
        if self.is_on:
            print(self.light_name + " is ON")
        else:
            print(self.light_name + " is OFF")

light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

light = SmartLight(light_name)

# perform action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action! Please enter ON or OFF.")

light.show_status()
```

#**2.  Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_idcard(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

employee = Employee(name, emp_id, department)
employee.display_idcard()
```

# **3.  Mobile Contact Record**

```python
class Contact_Record:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact_Record(name, phone)
contact.display_contact()
```

# **4. Product Price Tag Generator**

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + self.price)

name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)
product.display_price_tag()
```

# **5.  Movie Rating Display System**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)
movie.display_details()
```

# **6. Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(name, address)
delivery.display_delivery()
```

1. Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

# Split sentence into the words
words = sentence.split()

# Remove duplicates words using set
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

 2. Highest Salary from Employee Data

```python

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize with the first employee
employee_names = list(employees.keys())
highest_employee = employee_names[0]
highest_salary = employees[highest_employee]

# Compare each salary to find the highest
for name, salary in employees.items():
  if salary > highest_salary:
    highest_salary = salary
    highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

3. Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

max = numbers[0]
min = numbers[0]

# Compare each element with the min and max values
for num in numbers:
    if num > max:
        max = num
    if num < min:
        min = num

print("List:", numbers)
print("Max:", max)
print("Min:", min)
```

4. Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

counter = 0

for price in prices:
    if price > 1000:
        counter = counter + 1

print("Products above 1000: ", counter)
```

5. Calculate Attendance Percentage


```python

attendance = ["P", "P", "A", "P", "P"]

# Count number of present
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Total number of days
total_days = len(attendance)

# Calculate attendance percentage
attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

7. Count Character Frequency

```python
# Input string
text = "python"

# Dictionary to store character frequency
char_count = {}

# Loop through each character in the string
for ch in text:
    if ch in char_count:
        char_count[ch] += 1  # Increment count if already exists
    else:
        char_count[ch] = 1   # Initialize count if not exists

print(char_count)
```

 10. Calculate Average Marks

```python
marks = [70, 69, 80, 85]

total_marks = 0

for mark in marks:
    total_marks += mark

average_marks = total_marks / len(marks)

print("Average Marks:", average_marks)
```
